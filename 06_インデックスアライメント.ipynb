# レシピ49 別のDataFrameからカラムを追加

#### 従業員データセットにその従業員の部門最高給与のカラムを追加する

```python
# 新たなDataFrameにカラムDEPARTMRNTとBASE_SALARYを選ぶ
employee = pd.read_csv('data/employee.csv')
dept_sal = employee[['DEPARTMENT', 'BASE_SALARY']]
```

```python
# 部門内で給与をソートする
dept_sal = dept_sal.sort_values(['DEPARTMENT', 'BASE_SALARY'],ascending=[True, False])
```

```python
# drop_duplicatesメソッドを使い各部門の先頭行を残す
max_dept_sal = dept_sal.drop_duplicates(subset='DEPARTMENT')
max_dept_sal.head()
```

```python
# DEPARTMENTカラムをindexにする
max_dept_sal = max_dept_sal.set_index('DEPARTMENT')
employee = employee.set_index('DEPARTMENT')
```

```python
# employee DataFrameに新たなカラムを追加する
employee['MAX_DEPT_SALARY'] = max_dept_sal['BASE_SALARY']
employee.head()
```

```python
# BASE_SALARYがMAX_DEPT_SALARYより大きな行があるかqueryメソッドで確認
employee.query('BASE_SALARY > MAX_DEPT_SALARY')
```

```python
# sampleメソッドで10行を非復元無作為抽出する
np.random.seed(1234)
random_salary = dept_sal.sample(n=10).set_index('DEPARTMENT')
random_salary
```

```python
# indexに重複があると新たなカラムを追加しようとするとエラーとなる
employee['RANDOM_SALARY'] = random_salary['BASE_SALARY']
```

# レシピ51 メソッドチェイニングでidxmaxの代用

#### 基本的なメソッドをチェイニングして、カラムの最大値を含む行インデックス値を全て求める

```python
# レシピ50と同じ処理で数値カラムだけ得る
college = pd.read_csv('data/college.csv', index_col='INSTNM')
cols = ['MD_EARN_WNE_P10', 'GRAD_DEBT_MDN_SUPP']

for col in cols:
    college[col] = pd.to_numeric(college[col], errors='coerce')
    
college_n = college.select_dtypes(include=[np.number])
criteria = college_n.nunique() == 2
binary_cols = college_n.columns[criteria].tolist()
college_n = college_n.drop(labels=binary_cols, axis='columns')    
```

```python
# maxメソッドで各カラムの最大値を求める
college_n.max().head()
```

```python
# eqメソッドを使って値をそれぞれのカラムの最大値かテストする
college_n.eq(college_n.max()).head()
```

```python
# anyメソッドを使って少なくとも１つTrueを持つ行を全て見つける
has_row_max = college_n.eq(college_n.max()).any(axis='columns')
has_row_max.head()
```

```python
# has_row_maxの数を数える
college_n.shape
```

```python
has_row_max.sum()
```

```python
# sumsumメソッドを実行して全True値の累積和をとる
college_n.eq(college_n.max()).cumsum()
```

```python
# cumsumメソッドを再度チェイニングすると、値1は各カラムで1回しか現れずそれが最大値の先頭になる
college_n.eq(college_n.max()).cumsum().cumsum()
```

```python
# eqメソッドで各値が1と等しいか試験でき、それからanyメソッドを使い少なくとも1つのTrue値をもつ行を見つける
has_row_max2 = college_n.eq(college_n.max()) \
                        .cumsum() \
                        .cumsum() \
                        .eq(1) \
                        .any(axis='columns')
has_row_max2.head()                        
```

```python
# has_row_max2がカラム数より多くのTrue値を持たないことをチェックする
has_row_max2.sum()
```

```python
# Trueの全大学を抽出するのに、SeriesにBooleanインデックス法を使う
idxmax_cols = has_row_max2[has_row_max2].index
idxmax_cols
```

```python
# idxmaxメソッドで見つけたのと同じか確認する
set(college_n.idxmax().unique()) == set(idxmax_cols)
```

# レシピ52 最多の最大値を求める

#### idxmaxメソッドで各大学の学部生のうちでパーセントが最高の人種を見つけ、この最大値の分布を求める

```python
# 人種パーセントを含むカラムだけを選ぶ
college = pd.read_csv('data/college.csv', index_col='INSTNM')
college_ugds = college.filter(like='UGDS_')
college_ugds.head()
```

```python
# idxmaxメソッドを使い、各行で人種のパーセントが最多のカラム名を取得
highest_percentage_race = college_ugds.idxmax(axis='columns')
highest_percentage_race.head()
```

```python
# value_countsメソッドを使い、最多人種の分布を返す
highest_percentage_race.value_counts(normalize=True)
```

```python
# 黒人学生が一番多い大学で、次に多い人種の分布を調べる
college_black = college_ugds[highest_percentage_race == 'UGDS_BLACK']
college_black = college_black.drop('UGDS_BLACK', axis='columns')
college_black.idxmax(axis='columns').value_counts(normalize=True)
```
